```python
import pandas as pd
```

```python
PATH="https://raw.githubusercontent.com/underthelights/WebsiteFE/master/tangtang-revised.csv"
    
df = pd.read_csv(PATH)
df
```

```python
# 날짜 컬럼 생성
df['month'] = df['month'].astype(str)
df['year'] = df['year'].astype(str)
df['day'] = df['day'].astype(str)
df['month'] = df['month'].apply(lambda x: x.zfill(2) if int(x) < 10 else x)
df['Date'] = df['year'] + '-' + df['month'] + '-' + df['day']
df
```

```python
# # 'percent' 컬럼의 값을 리스트에서 소수점 형태로 변환
# df['percent'] = df['percent'].apply(lambda x: float(x[0]))
import ast
df['percent'] = df['percent'].apply(lambda x: float(ast.literal_eval(x)[0]))
df
```

```python
# 'sentiment' 컬럼이 '부정'인 경우 'percent' 값을 마이너스로 변경
df.loc[df['sentiment'] == '부정', 'percent'] = df.loc[df['sentiment'] == '부정', 'percent'] * -1

df
```

```python
df.to_csv('../file/tangtang-revised.csv')
```

```python
import pandas as pd 
PATH="https://raw.githubusercontent.com/underthelights/WebsiteFE/master/tangtang-revised.csv"
    
df = pd.read_csv(PATH)
df
```

```python
from collections import Counter
from konlpy.tag import Okt
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load tokenizer (e.g., Okt)
tokenizer = Okt()

# Define function to preprocess text
def preprocess_text(text):
    tokens = tokenizer.pos(text, stem=True, norm=True)
    filtered_tokens = [word for word, pos in tokens if pos in ['Noun', 'Adjective', 'Verb']]
    return filtered_tokens


# Concatenate all reviews
all_reviews = ' '.join(df['reviews'])
```

```python

    # Preprocess the text and count the keywords
keywords = preprocess_text(all_reviews)
keyword_counts = Counter(keywords)
```

```python
# Create treemap
fig, ax = plt.subplots(figsize=(8, 8))
wordcloud = WordCloud(width=800, height=800,
                    background_color='white',
                    max_words=50).generate_from_frequencies(keyword_counts)
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis('off')
```
