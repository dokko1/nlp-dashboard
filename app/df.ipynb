```python
import pandas as pd
```

```python
PATH="https://raw.githubusercontent.com/underthelights/WebsiteFE/master/tangtang-revised.csv"
    
df = pd.read_csv(PATH)
df
```

```python
# 날짜 컬럼 생성
df['month'] = df['month'].astype(str)
df['year'] = df['year'].astype(str)
df['day'] = df['day'].astype(str)
df['month'] = df['month'].apply(lambda x: x.zfill(2) if int(x) < 10 else x)
df['Date'] = df['year'] + '-' + df['month'] + '-' + df['day']
df
```

```python
# # 'percent' 컬럼의 값을 리스트에서 소수점 형태로 변환
# df['percent'] = df['percent'].apply(lambda x: float(x[0]))
import ast
df['percent'] = df['percent'].apply(lambda x: float(ast.literal_eval(x)[0]))
df
```

```python
# 'sentiment' 컬럼이 '부정'인 경우 'percent' 값을 마이너스로 변경
df.loc[df['sentiment'] == '부정', 'percent'] = df.loc[df['sentiment'] == '부정', 'percent'] * -1

df
```

```python
df.to_csv('../file/tangtang-revised.csv')
```
